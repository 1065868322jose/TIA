# MODELO DEL PORTAFOLIO DE MARKOWIZT

El Modelo de Markowitz, también conocido como Teoría Moderna de Portafolios, fue desarrollado por el economista Harry Markowitz en la década de 1950. Este modelo es fundamental en la teoría financiera y se utiliza para construir carteras óptimas de activos financieros.

En términos simples, el Modelo de Markowitz ayuda a los inversores a tomar decisiones sobre cómo distribuir su capital entre diferentes activos para maximizar el rendimiento esperado mientras se minimiza el riesgo. Se basa en la premisa de que los inversores son aversos al riesgo y desean maximizar la rentabilidad de su inversión.

El modelo tiene en cuenta dos aspectos principales:

+ Rendimiento esperado: Es la ganancia que se espera obtener de una inversión. Se calcula como el promedio ponderado de los rendimientos históricos de los activos en el portafolio.
+ Riesgo: Se refiere a la volatilidad o variabilidad de los rendimientos de los activos en el portafolio. El riesgo se mide típicamente mediante la desviación estándar de los rendimientos históricos.

El objetivo del Modelo de Markowitz es encontrar la combinación óptima de activos que maximice el rendimiento esperado dado un nivel de riesgo tolerado por el inversor, o minimice el riesgo dado un nivel de rendimiento esperado deseado.

## Eliminar las columnas "FECHA" del DataFrame df y crear uno nuevo llamado dfprecovid

```python
# Eliminar las columnas "FECHA" y "COLCAP" del DataFrame df
dfprecovid = df.drop(columns=['FECHA'])

# Verificar el nuevo DataFrame dfprecovid
print(dfprecovid.head())
```

## Calcular el rendimiento diario

```python
import numpy as np

# Calcular el rendimiento diario utilizando logaritmos para el DataFrame dfprecovid
dfprecovid_rendimientos = pd.DataFrame()
for col in dfprecovid.columns:
    dfprecovid_rendimientos[col] = np.log(dfprecovid[col] / dfprecovid[col].shift(1))

# Eliminar la primera fila que contendrá NaN
dfprecovid_rendimientos = dfprecovid_rendimientos.dropna()

# Mostrar las primeras filas del nuevo DataFrame
print(dfprecovid_rendimientos.head())
```

## El rendimiento anualizado

```python
# Calcular el rendimiento anualizado utilizando los rendimientos diarios
rendimiento_anualizado = dfprecovid_rendimientos.mean() * 252

# Mostrar los rendimientos anualizados
print(rendimiento_anualizado)
```

## El rendimiento anualizado Porcentual

```python
# Calcular el rendimiento anualizado utilizando los rendimientos diarios
rendimiento_anualizado = dfprecovid_rendimientos.mean() * 252 * 100

# Mostrar los rendimientos anualizados en forma porcentual
print(rendimiento_anualizado)
```

## desviación estándar anualizada

```python
# Calcular la desviación estándar de cada columna de rendimientos diarios
desviacion_estandar = dfprecovid_rendimientos.std()

# Ajustar la desviación estándar para un año completo (252 días)
desviacion_anualizada = desviacion_estandar * (252 ** 0.5)

# Mostrar la desviación estándar anualizada de cada columna
print(desviacion_anualizada)

```

## coeficiente de variación

```python
# Calcular el coeficiente de variación para cada acción
coeficiente_variacion = desviacion_anualizada / rendimiento_anualizado

# Mostrar el coeficiente de variación para cada acción
print(coeficiente_variacion)
```

## BETAS

```python
from scipy.stats import linregress

# Calcular la pendiente para cada columna
pendientes = {}
for columna in dfprecovid_rendimientos.columns:
    slope, intercept, r_value, p_value, std_err = linregress(dfprecovid_rendimientos['COLCAP'], dfprecovid_rendimientos[columna])
    pendientes[columna] = slope

# Mostrar las pendientes de cada columna
for columna, pendiente in pendientes.items():
    print(f"BETA {columna}: {pendiente}")
```

## ALFAS

```python
from scipy.stats import linregress

# Calcular la pendiente y la intersección para cada columna
resultados_regresion = {}
for columna in dfprecovid_rendimientos.columns:
    slope, intercept, r_value, p_value, std_err = linregress(dfprecovid_rendimientos['COLCAP'], dfprecovid_rendimientos[columna])
    resultados_regresion[columna] = intercept

# Mostrar las intersecciones de cada columna
for columna, interseccion in resultados_regresion.items():
    print(f"ALFA {columna}: {interseccion}")
```

## RIESGO SISTEMATICO

```python
# Calcular el riesgo sistemático para cada acción
resultados_riesgo_sistematico = {}
for columna in dfprecovid_rendimientos.columns:
    if columna != 'COLCAP':
        beta = pendientes[columna]  # Utilizamos las pendientes calculadas previamente
        desviacion = dfprecovid_rendimientos[columna].std() * np.sqrt(252)  # Calculamos la desviación estándar anualizada

        # Calcular riesgo sistemático (en porcentaje)
        riesgo_sistematico = (beta ** 2) * (desviacion ** 2) * 100

        resultados_riesgo_sistematico[columna] = riesgo_sistematico

# Riesgo sistemático para COLCAP es 100%
resultados_riesgo_sistematico['COLCAP'] = 100.0

# Mostrar los resultados de riesgo sistemático para cada acción
for columna, riesgo_sistematico in resultados_riesgo_sistematico.items():
    print(f"Riesgo sistemático para {columna}: {riesgo_sistematico:.2f}%")
```

## RIESGO NO SISTEMATICO

```python
# Calcular el riesgo no sistemático para cada acción
resultados_riesgo_no_sistematico = {}
for columna, riesgo_sistematico in resultados_riesgo_sistematico.items():
    riesgo_no_sistematico = 100 - riesgo_sistematico
    resultados_riesgo_no_sistematico[columna] = riesgo_no_sistematico

# Mostrar los resultados de riesgo no sistemático para cada acción
for columna, riesgo_no_sistematico in resultados_riesgo_no_sistematico.items():
    print(f"Riesgo no sistemático para {columna}: {riesgo_no_sistematico:.2f}%")
```

## Índice de Sharpe  |  Índice de Treynor

```python
import numpy as np

# Tasa libre de riesgo (supongamos que es 3%)
tasa_libre_riesgo = 0.0463

# Calcular el rendimiento anualizado para cada acción
rendimientos_anualizados = {}
for columna in dfprecovid_rendimientos.columns:
    rendimiento_anualizado = dfprecovid_rendimientos[columna].mean() * 252  # Promedio de rendimientos diarios multiplicado por 252 días
    rendimientos_anualizados[columna] = rendimiento_anualizado

# Calcular la desviación estándar anualizada para cada acción
desviaciones_estandar_anualizadas = dfprecovid_rendimientos.std() * np.sqrt(252)

# Calcular el índice de Sharpe y el índice de Treynor para cada acción
indices_sharpe = {}
indices_treynor = {}

for columna in dfprecovid_rendimientos.columns:
    rendimiento_anualizado = rendimientos_anualizados[columna]
    desviacion_anualizada = desviaciones_estandar_anualizadas[columna]
    beta = pendientes[columna]

    # Índice de Sharpe
    sharpe = (rendimiento_anualizado - tasa_libre_riesgo) / desviacion_anualizada

    # Índice de Treynor
    treynor = (rendimiento_anualizado - tasa_libre_riesgo) / beta

    indices_sharpe[columna] = sharpe
    indices_treynor[columna] = treynor

# Mostrar los resultados de los índices de Sharpe y Treynor para cada acción en tres columnas
print("Acción     |  Índice de Sharpe  |  Índice de Treynor")
print("-" * 55)
for columna, sharpe in indices_sharpe.items():
    treynor = indices_treynor[columna]
    print(f"{columna:<11} |  {sharpe:>18.4f}  |  {treynor:>16.4f}")
```

## Acciones con Sharpe y Treynor positivos

```python
import pandas as pd

# Crear un diccionario con los datos de los índices de Sharpe y Treynor
datos = {
    'Sharpe': indices_sharpe,
    'Treynor': indices_treynor
}

# Convertir el diccionario en un DataFrame
df_indices = pd.DataFrame(datos)

# Filtrar las acciones con Sharpe y Treynor positivos
acciones_positivas = df_indices[(df_indices['Sharpe'] > 0) & (df_indices['Treynor'] > 0)]

# Mostrar el DataFrame resultante
print(acciones_positivas)
```

## Portafolio de Markowitz

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Calcular rendimientos esperados y matriz de covarianza
rendimientos_esperados = portafolio_prepandemia.mean()
matriz_covarianza = portafolio_prepandemia.cov()

# Función objetivo para minimizar la volatilidad del portafolio
def volatilidad_portafolio(weights):
    return np.sqrt(np.dot(weights.T, np.dot(matriz_covarianza, weights)))

# Restricción: la suma de los pesos debe ser igual a 1
def constraint(weights):
    return np.sum(weights) - 1

# Inicializar los pesos de manera uniforme
n_acciones = len(rendimientos_esperados)
weights_inicial = np.array([1 / n_acciones] * n_acciones)

# Definir el peso mínimo como el 3%
peso_minimo = 0.03

# Definir los límites de los pesos (entre el peso mínimo y 1)
limites_pesos = tuple((peso_minimo, 1) for _ in range(n_acciones))

# Optimización para encontrar el portafolio de Markowitz
optimizacion = minimize(volatilidad_portafolio, weights_inicial, method='SLSQP', constraints={'type':'eq', 'fun': constraint}, bounds=limites_pesos)
pesos_optimos = optimizacion.x

# Calcular rendimiento y volatilidad del portafolio óptimo
rendimiento_optimo = np.dot(rendimientos_esperados, pesos_optimos)
volatilidad_optima = volatilidad_portafolio(pesos_optimos)

# Calcular índice de Sharpe
tasa_libre_riesgo = 0.0463
sharpe_ratio = (rendimiento_optimo - tasa_libre_riesgo) / volatilidad_optima

# Convertir pesos óptimos a porcentaje y agregar nombres de acciones
pesos_optimos_porcentaje = (pesos_optimos * 100).round(2)
acciones = portafolio_prepandemia.columns
portafolio_optimo_df = pd.DataFrame({'Acción': acciones, 'Peso Óptimo (%)': pesos_optimos_porcentaje})

# Imprimir resultados
print("Pesos óptimos:")
print(portafolio_optimo_df)
print("Rendimiento esperado del portafolio óptimo:", f"{rendimiento_optimo:.4f}")
print("Volatilidad del portafolio óptimo:", f"{volatilidad_optima:.4f}")
print("Índice de Sharpe del portafolio óptimo:", sharpe_ratio)

# Generar una variedad de asignaciones de pesos para cada acción en el portafolio
num_portafolios = 100000
portafolios = []

for i in range(num_portafolios):
    weights = np.random.random(n_acciones)
    weights /= np.sum(weights)
    rendimiento = np.dot(rendimientos_esperados, weights)
    volatilidad = volatilidad_portafolio(weights)
    sharpe_ratio = (rendimiento - tasa_libre_riesgo) / volatilidad
    portafolios.append([rendimiento, volatilidad, sharpe_ratio])

portafolios = np.array(portafolios)

# Graficar la frontera eficiente
plt.figure(figsize=(10, 6))
plt.scatter(portafolios[:, 1], portafolios[:, 0], c=portafolios[:, 2], cmap='viridis')
plt.title('Frontera Eficiente')
plt.xlabel('Volatilidad')
plt.ylabel('Rendimiento')
plt.colorbar(label='Índice de Sharpe')
plt.scatter(volatilidad_optima, rendimiento_optimo, marker='*', color='r', s=100, label='Portafolio Óptimo')
plt.legend()
plt.show()
```
